```python
import pandas as pd
import numpy as np
import seaborn as sns
from sklearn.preprocessing import StandardScaler, OneHotEncoder
```

```python
df_train = pd.read_csv('D:/Data/tmlcc-2021/train_clean_no_outlier_01-10.csv')
df_feat_train = pd.read_csv('D:/Code/tcml/train_mnc_metal.csv')
df_pretest = pd.read_csv('D:/Data/tmlcc-2021/test_set_clean_04-10.csv')
df_feat_pretest = pd.read_csv('D:/Code/tcml/test_mnc_metal.csv')
```

```python
df_fix = pd.read_csv('D:/Data/tmlcc-2021/sra_2to15.csv')
df_fix_train = df_fix[df_fix['flag']=='train']
df_fix_test = df_fix[df_fix['flag']=='test']
```

```python
df_fix_train
```

```python
for i in range(len(df_fix_train)):
    name = df_fix_train.iloc[i]['MOFname']
    idx = df_train.loc[df_train['MOFname']==name].index.item()
    
    df_train.at[idx,'metal_linker'] = df_fix_train.iloc[i]['metal_linker']
    df_train.at[idx,'organic_linker1'] = df_fix_train.iloc[i]['organic_linker1']
    df_train.at[idx,'organic_linker2'] = df_fix_train.iloc[i]['organic_linker2']
```

```python
for i in range(len(df_fix_test)):
    name = df_fix_test.iloc[i]['MOFname']
    idx = df_pretest.loc[df_pretest['MOFname']==name].index.item()
    
    df_pretest.at[idx,'metal_linker'] = df_fix_test.iloc[i]['metal_linker']
    df_pretest.at[idx,'organic_linker1'] = df_fix_test.iloc[i]['organic_linker1']
    df_pretest.at[idx,'organic_linker2'] = df_fix_test.iloc[i]['organic_linker2']
```

## bag of word functional group , one-hot topology

### topology

```python
enc_topology = OneHotEncoder(handle_unknown='ignore')
```

```python
enc_topology.fit(df_train[['topology']].values)
```

```python
enc_topology.categories_
```

### train

```python
train_topology = enc_topology.transform(df_train[['topology']].values).toarray()
df_train_topology = pd.DataFrame(train_topology,columns = enc_topology.categories_)
```

### pretest

```python
pretest_topology = enc_topology.transform(df_pretest[['topology']].values).toarray()
df_pretest_topology = pd.DataFrame(pretest_topology,columns = enc_topology.categories_)
```

```python
df_train_topology
```

### bag of word functional group

```python
enc_func = OneHotEncoder(handle_unknown='ignore')
```

```python
enc_func.fit(df_train[['functional_groups 1']].values)
enc_func.categories_
```

### train

```python
train_func_1 = enc_func.transform(df_train[['functional_groups 1']].values).toarray()
train_func_2 = enc_func.transform(df_train[['functional_groups 2']].values).toarray()
train_func = train_func_1+train_func_2
df_train_func = pd.DataFrame(train_func,columns = enc_func.categories_)
```

```python
df_train_func
```

```python
pretest_func_1 = enc_func.transform(df_pretest[['functional_groups 1']].values).toarray()
pretest_func_2 = enc_func.transform(df_pretest[['functional_groups 2']].values).toarray()
pretest_func = pretest_func_1+pretest_func_2
df_pretest_func = pd.DataFrame(pretest_func,columns = enc_func.categories_)
```

```python
df_pretest_func
```

```python
df_train = df_train.reset_index(drop = True).join(df_train_topology)
df_train = df_train.join(df_train_func)
#df_train = pd.concat([df_train,df_train_topology,df_train_func],ignore_index=True,axis = 1)
```

# normalize

## scaler sklearn (z score) (2)


```python
target_train = [col for col in df_train.columns if col not in ['CO2_working_capacity [mL/g]','MOFname']]
target_pretest = [col for col in df_pretest.columns if col not in ['MOFname']]
```

```python
for i in range(len(target_train)):
    x_train = df_train[[target_train[i]]].values
    x_pretest = df_pretest[[target_pretest[i]]].values
    
    scaler = StandardScaler()    
    scaler.fit(x_train)
    x_train = scaler.transform(x_train)
    x_pretest = scaler.transform(x_pretest)
    
    df_train[target_train[i]] = list(x_train.squeeze(1))
    df_pretest[target_pretest[i]] = list(x_pretest.squeeze(1))
```
